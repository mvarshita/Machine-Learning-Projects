# Decision Tree Regression on Cars Dataset

**Import the required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb 
```

**Read the data set**

```python
carsDataSet = pd.read_csv("Cars93.csv")
carsDataSet.head(3)
```

**Drop the first column as it is unwanted**

```python
cars=carsDataSet.drop(labels ="Unnamed: 0" ,axis = 1)
cars.head(3)
```

```python
cars.info()
```

**Data Pre-processing**

```python
cars.isnull().sum()
```

**Exploratory Data Analysis(EDA)**

```python
#seggregating categorical and continuous variables into its corresponsing lists for 
#further data visualization.

categoricalColumnsList = []
continuousColumnsList = []

def catcon(dataset):
    for colNames in dataset.columns:
        if dataset[colNames].dtypes =="object":
            categoricalColumnsList.append(colNames)
        else:
            continuousColumnsList.append(colNames)
    return categoricalColumnsList,continuousColumnsList
```

```python
categorical,continuous = catcon(cars)
```

```python
print("Categorical columns : %s"%categorical)
print("                                 ")
print("Continuous columns : %s"%continuous)
```

```python
plt.figure(figsize=(15,27))
j= 1
for catColumns in categorical:
    plt.subplot(7,3,j)
    sb.boxplot(cars[catColumns],cars.Price)
    plt.xlabel("%s"%catColumns)
    plt.ylabel("Price")
    j = j +1
plt.show()
```

```python
plt.figure(figsize=(15,27))
j= 1
for contColumns in continuous:
    plt.subplot(7,3,j)
    plt.scatter(cars[contColumns],cars.Price)
    plt.xlabel("%s"%contColumns)
    plt.ylabel("Price")
    j = j +1
plt.show()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cars=cars.apply(le.fit_transform)
```

**Prepare the data for sampling**

```python
X = cars[cars.columns[~cars.columns.isin(['Price'])]]
Y = cars[["Price"]]
```

**Split the data into training and testing sets**

```python
from sklearn.model_selection import train_test_split

# splits the training and test data set in 80% : 20%
# assign random_state to any value.This ensures consistency.
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2, random_state=24)
```

```python
#check if the sampling done is good or bad

sb.distplot(Y,color = "green")
sb.distplot(ytrain,color = "purple")
sb.distplot(ytest,color = "orange")
```

**Observation : The sampling done is good.**

**Create a dictionary of parameters to be tuned **

```python
tp = {"max_leaf_nodes":range(2,100)}
```

**Create the model using sklearn DecisionTreeRegressor**

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state= 40)
```

**Find the best value of tuning parameter to reduce the model error using cross validation**

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(dtr,tp,scoring="neg_mean_squared_error",cv=4)
cvmodel = cv.fit(xtrain,ytrain.values.ravel())
dtrBestParams=cvmodel.best_params_
print("The best parameters for Decision Tree Regressor are %s  "%dtrBestParams)
```

```python
dtr = DecisionTreeRegressor(random_state=40,max_leaf_nodes= 16)
modelDecisionTreeRegressor = dtr.fit(xtrain,ytrain.values.ravel())
```

**Build Predictions**

```python
predDecisionTreeRegressor=modelDecisionTreeRegressor.predict(xtest)
```

**Confusion matrix and Mean Absolute Error for Decision Tree Regression**

```python
from sklearn.metrics import mean_absolute_error

print("The mean absolute error for Decision Tree Regression is %.2f"%mean_absolute_error(ytest,predDecisionTreeRegressor))
```
