# Random Forest Classification on Cars Dataset

**Import the required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb 
```

**Read the data set**

```python
carsDataSet = pd.read_csv("Cars93.csv")
carsDataSet.head(3)
```

**Drop the first column as it is unwanted**

```python
cars=carsDataSet.drop(labels ="Unnamed: 0" ,axis = 1)
cars.head(3)
```

```python
cars.info()
```

**Data Pre-processing**

```python
cars.isnull().sum()
```

**Exploratory Data Analysis(EDA)**

```python
#seggregating categorical and continuous variables into its corresponsing lists for 
#further data visualization.

categoricalColumnsList = []
continuousColumnsList = []

def catcon(dataset):
    for colNames in dataset.columns:
        if dataset[colNames].dtypes =="object":
            categoricalColumnsList.append(colNames)
        else:
            continuousColumnsList.append(colNames)
    return categoricalColumnsList,continuousColumnsList
```

```python
categorical,continuous = catcon(cars)
```

```python
print("Categorical columns : %s"%categorical)
print("                                 ")
print("Continuous columns : %s"%continuous)
```

```python
plt.figure(figsize=(15,27))
j= 1
for contColumns in continuous:
    plt.subplot(7,3,j)
    sb.boxplot(cars.Type,cars[contColumns])
    plt.xlabel("Type")
    plt.ylabel("%s"%contColumns)
    j = j +1
plt.show()
```

**Identify the unique elements of the categorical data variable "Type"**

```python
print(cars['Type'].unique())
```

**Count the number of unique elements of the categorical data variable "Type"**

```python
cars['Type'].value_counts()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cars=cars.apply(le.fit_transform)
```

**Prepare the data for sampling**

```python
X = cars[cars.columns[~cars.columns.isin(["Type"])]]
Y = cars[["Type"]]
```

**Split the data into training and testing sets**

```python
from sklearn.model_selection import train_test_split

# splits the training and test data set in 80% : 20%
# assign random_state to any value.This ensures consistency.
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2, random_state=24)
```

```python
#check if the sampling done is good or bad

sb.distplot(Y,color = "green")
sb.distplot(ytrain,color = "purple")
sb.distplot(ytest,color = "orange")
```

**Observation : The sampling done is good.**

**Create a dictionary of parameters to be tuned **

```python
tp = {"n_estimators":range(2,50)}
```

**Create the model using sklearn RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state= 30)
```

**Find the best value of tuning parameter to reduce the model error using cross validation**

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(rfc,tp,scoring="accuracy",cv=4)
cvmodel = cv.fit(xtrain,ytrain.values.ravel())
rfcBestParams=cvmodel.best_params_
print("The best parameters for Random Forest Classifier are %s  "%rfcBestParams)
```

```python
rfc = RandomForestClassifier(random_state=30,n_estimators=30)
modelRandomForestClassifier=rfc.fit(xtrain,ytrain.values.ravel())
```

**Build Predictions**

```python
predRandomForestClassifier=modelRandomForestClassifier.predict(xtest)
```

**Confusion matrix and Accuracy score for Random Forest Classification**

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cmRandomForestClassifier=confusion_matrix(ytest,predRandomForestClassifier)

xAxisLabels = [' Small ',' Midsize ',' Compact ',' Large ',' Sporty ',' Van '] # labels for x-axis
yAxisLabels = [' Small ',' Midsize ',' Compact ',' Large ',' Sporty ',' Van '] # labels for y-axis

# create seaborn heatmap with required labels
sb.heatmap(cmRandomForestClassifier, annot=True,cmap = "YlGnBu",xticklabels=xAxisLabels, yticklabels=yAxisLabels)

print("The accuracy score for Random Forest Classification is %.2f"%accuracy_score(ytest,predRandomForestClassifier))
```

**Calculate the feature importance matrix **

```python
modelRandomForestClassifier.feature_importances_
```

**Print the name and its importance of each feature in a ascending manner**

```python
featureImportanceDf = pd.DataFrame(list(zip(X.columns,modelRandomForestClassifier.feature_importances_)),columns=["features","importance"])

#set index from column Source
featureImportanceDf = featureImportanceDf.set_index('features').rename_axis(None)

#sorting the obtained features based on theri importance
sortedFeatureImportance = featureImportanceDf.sort_values(by=["importance"])
print(sortedFeatureImportance)
```

```python
# Plotting the feature importance 
sortedFeatureImportance.plot.barh(figsize=(10,10),color = "green")
plt.show()
```
